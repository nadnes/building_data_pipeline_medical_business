Standardizing Formats


```python
data
```

```python
# Convert 'forward_msg_from_peer_id' column from scientific notation to float64
data['forward_msg_from_peer_id'] = data['forward_msg_from_peer_id'].astype('float64')

print("\nDataFrame after converting to float64:")
print(data)
```

```python
# Convert 'Timestamp' column to datetime format
data['date'] = pd.to_datetime(data['date'])

print("\nDataFrame after converting to datetime:")
print(data)
```

Data Validation

```python
data['date'] = pd.to_datetime(data['date'], errors='coerce') 
# Check for missing values
missing_values = data.isnull().sum()

print("Validation Report:")
print("Missing Values:")
print(missing_values)

print("\nDataFrame after validation:")
print(data)
```

```python
data['msg_from_peer'] = pd.to_numeric(data['msg_from_peer'], errors='coerce') 
# Check for missing values
missing_values = data.isnull().sum()

print("Validation Report:")
print("Missing Values:")
print(missing_values)

print("\nDataFrame after validation:")
print(data)
```
